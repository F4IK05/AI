```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
hours = [0, 4, 8, 12, 16, 20, 24]
temperature = [15, 14, 18, 22, 25, 20, 17]
```

```python
plt.plot(hours, temperature)
plt.xlabel("Часы")
plt.ylabel("Температура")
plt.title("Температура в течение суток")
plt.grid(True)
plt.show()
```

```python
X = np.array(hours).reshape(-1, 1)
y = np.array(temperature)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Мало данных поэтому такие результаты
print("MAE:", mae)
print("MSE:", mse)
print("R2:", r2)
```

```python
new_h = np.array([3]).reshape(-1, 1)
pred_t = model.predict(new_h)

print(f"{pred_t[0]:.2f}")
```
