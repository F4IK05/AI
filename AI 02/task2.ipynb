```python
import numpy as np
from tabulate import tabulate
```

```python
arr = np.arange(15).reshape(3, 5)

# Выведите исходный массив
print("Исходный массив:\n")
table = tabulate(arr, tablefmt="fancy_grid")
print(table)
```

```python
# Выполните транспонирование массива
transpose_arr = np.transpose(arr)

print("Траспонированный массив:\n")
table = tabulate(transpose_arr, tablefmt="fancy_grid")
print(table)
```

```python
reshape_arr = np.reshape(arr, (5, 3))

print("Массив после reshape:\n")
table = tabulate(reshape_arr, tablefmt="fancy_grid")
print(table)
```
