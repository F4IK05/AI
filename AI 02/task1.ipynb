```python
import numpy as np
from tabulate import tabulate
```

```python
arr = np.random.randint(1, 50, size=(4, 4))
```

```python
# Выведите массив в читаемом виде (в виде таблицы).
table = tabulate(arr, tablefmt="fancy_grid")
print(table)
```

```python
# Найдите сумму элементов по каждой строке и по каждому столбцу (используйте параметр axis).
sum_rows = np.sum(arr, axis=1)
sum_cols = np.sum(arr, axis=0)

print("Суммы по строкам:")
for i in range(len(sum_rows)):
    print(f"\tСумма строки {arr[i]} = {sum_rows[i]}")

print("\nСуммы по столбцам:")
for i in range(len(sum_cols)):
    print(f"\tСумма столбца {arr[:, i]} = {sum_cols[i]}")

# Определите общее среднее значение массива.
print("\nОбщее среднее значение: ", np.average(arr))
```
