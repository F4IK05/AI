```python
import numpy as np
from IPython.display import Markdown, display
```

```python
sales = [120, 135, 100, 150, 90, 200, 170]

total_sales = np.sum(sales)
avg = np.mean(sales)
max_val = np.max(sales)
min_val = np.min(sales)

display(Markdown(f"""
# Отчёт по продажам

| Показатель | Значение |
|------------|---------------|
| Сумма    | {total_sales} |
| Среднее  | {avg:.2f} |
| Максимум | {max_val} |
| Минимум  | {min_val} |
"""))
```
