```python
import pandas as pd
```

```python
df = pd.read_csv("data/orders.csv")
```

```python
# Выведите первые десять строк.

first_10 = df.head(10)
print(first_10)
```

```python
# Определите количество заказов каждого клиента.

count_orders = df["Клиент"].value_counts()
print(count_orders)
```

```python
# Найдите максимальную и минимальную суммы заказов.

max_sum = df["Сумма"].max()
min_sum = df["Сумма"].min()

print(max_sum)
print(min_sum)
```

```python
# Подсчитайте общую сумму всех заказов.

total_sum = df["Сумма"].sum()

print(total_sum)
```
