```python
import pandas as pd
```

```python
df = pd.read_csv("data/products.csv")

print(df)
```

```python
# Найдите все продукты с калорийностью выше 300.

high_cal = df[df["Калории"] > 300]
print(high_cal)
```

```python
# Подсчитайте среднее количество белков по категориям.

avg_protein = df.groupby("Категория")["Белки"].mean()
print(avg_protein)
```

```python
# Отсортируйте DataFrame по калорийности в порядке убывания.

product_filter = df.sort_values(by="Калории", ascending=False)
print(product_filter)
```
