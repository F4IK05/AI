```python
import pandas as pd
```

```python
df = pd.read_csv("data/employees.csv")

print(df)
```

```python
# Подсчитайте общее количество часов по каждому сотруднику.

hours_by_emp = df.groupby("Имя")["Часы"].sum()
print(hours_by_emp)
```

```python
# Подсчитайте общее количество часов по каждому проекту.

hours_by_project = df.groupby("Проект")["Часы"].sum()
print(hours_by_project)
```

```python
# Найдите сотрудника, затратившего больше всего часов.

best_emp = df.groupby("Имя")["Часы"].sum().idxmax()
print(best_emp)
```
