```python
import pandas as pd
```

```python
df = pd.read_csv("data/tickets.csv")

print(df)
```

```python
# Подсчитайте общее количество проданных билетов по каждому фильму.

tickets_sold = df.groupby("Фильм")["Проданные билеты"].sum()
print(tickets_sold)
```

```python
# Подсчитайте общее количество билетов по каждому городу.

tickets_by_city = df.groupby("Город")["Проданные билеты"].sum()
print(tickets_by_city)
```

```python
# Найдите фильм с наибольшим количеством продаж.

top_film = df.groupby("Фильм")["Проданные билеты"].sum().idxmax()
print(top_film)
```
