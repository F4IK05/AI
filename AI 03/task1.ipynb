```python
import pandas as pd
```

```python
data = {
    "Название": [
        "Мастер и Маргарита",
        "1984",
        "Преступление и наказание",
        "Гарри Поттер",
        "Война и мир",
        "Маленький принц",
        "Атлант расправил плечи"
    ],
    "Автор": [
        "Михаил Булгаков",
        "Джордж Оруэлл",
        "Фёдор Достоевский",
        "Дж. К. Роулинг",
        "Лев Толстой",
        "Антуан де Сент-Экзюпери",
        "Айн Рэнд"
    ],
    "Год издания": [2010, 2017, 2019, 2016, 2012, 2021, 2018],
    "Цена": [450, 380, 520, 670, 890, 310, 750]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
# Определите среднюю цену книг.

mean_price = df["Цена"].mean()
print(mean_price)
```

```python
# Найдите книги, изданные после 2015 года.

book_filter_1 = df[df["Год издания"] > 2015]
print(book_filter_1)
```

```python
# Отсортируйте DataFrame по цене в порядке возрастания.

book_filter_2 = df.sort_values(by="Цена", ascending=True)
print(book_filter_2)
```
